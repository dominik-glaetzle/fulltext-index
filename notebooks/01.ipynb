```python
import sys
import pandas as pd
sys.path.append("../src")


from read import read_texts_from_folder
from tokenizer import tokenize
from analyzer import analyze_tokens, get_top_terms
from utils import get_stopwords
from utils import merge_texts
```

## Load data + preprocessing

```python
texts = read_texts_from_folder("../data")
full_text = merge_texts(texts)
all_tokens = tokenize(full_text)
print(f"{len(all_tokens)} loaded")
```

## Analysis

```python
stopwords = get_stopwords()
stats = analyze_tokens(all_tokens, stopwords)
```

## Stats

```python
print("Total terms: ", stats["total_terms"])
print("Unique terms: ", stats["unique_terms"])
print("Stopword count: ", stats["stopword_count"])
```

## Top 50

```python
top_50_without = get_top_terms(stats["term_freq"], stopwords_set=stopwords)
top_50= get_top_terms(stats["term_freq"])

df = pd.DataFrame(top_50, columns=["Word", "Frequency"])
df_without = pd.DataFrame(top_50_without, columns=["Word (without stopwords)", "Frequency"])
df.index += 1
df_without.index += 1

df_combined = pd.concat([df, df_without], axis=1)
df_combined
```
